```python
import numpy as np
import pandas as pd
```

```python
import pickle
matches = pickle.load(open('dataset_1.pkl' , 'rb'))
```

```python
matches
```

```python
matches.info()
```

```python
count = 1
delivery_df1 = pd.DataFrame()
for index, row in matches.iterrows():
    count+=1
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_of_dismissed = []
    teams = []
    batting_team = []
    match_id = []
    city = []
    venue = []

    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            batting_team.append(row['innings'][0]['1st innings']['team'])
            teams.append(row['info.teams'])
            ball_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            try:
                player_of_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                player_of_dismissed.append('0')
            match_id.append(index + 1)       
               
    loop_df = pd.DataFrame({
            'match_id':match_id,
            'teams':teams,
            'batting_team':batting_team,
            'ball':ball_of_match,
            'batsman':batsman,
            'bowler':bowler,
            'runs':runs,
            'player_dismissed':player_of_dismissed,
            'city':city,
            'venue':venue
        })
    delivery_df1 = pd.concat([delivery_df1, loop_df], ignore_index=True)
```

```python
df = matches[["info.outcome.winner" , "match_id"]].reset_index().merge(delivery_df1 , on= "match_id").drop("index" , axis = 1)
```

```python
df
```

```python
def bowl(row):
    for team in row["teams"]:
        if team != row["batting_team"]:
            return team
```

```python
df["bowling_team"] = df.apply(bowl , axis=1)
```

```python
df.drop("teams" , axis =1 , inplace =True)
```

```python
teams = ["Australia",
        "Pakistan",
        "New Zealand",
        "Bangladesh",
        "South Africa",
        "India",
        "England",
        "Sri Lanka",
        "Netherlands"]
```

```python
df = df[df["batting_team"].isin(teams)]
df = df[df["bowling_team"].isin(teams)]
```

```python
df[df["bowling_team"] == "Afghanisthan"]
```

```python
df["runs"] = pd.to_numeric(df["runs"])
df["current_score"] = df.groupby("match_id")["runs"].cumsum()
```

```python
df
```

```python
df["overs"] = df["ball"].apply(lambda x : str(x).split(".")[0])
df["ball_no"] = df["ball"].apply(lambda x : str(x).split(".")[1])
```

```python
df["balls_played"] = df["overs"].astype(int) * 6 + df["ball_no"].astype(int)
```

```python
df
```

```python
df["balls_left"] = 300 - df["balls_played"]
df["balls_left"] = df["balls_left"].apply(lambda x:0 if x<0 else x)
```

```python
df
```

```python
def result(row):
    return 1 if row['batting_team'] == row['info.outcome.winner'] else 0
```

```python
df['result'] = df.apply(result,axis=1)
```

```python
df["player_dismissed"] = df["player_dismissed"].apply(lambda x:0 if x=="0" else 1)
df["player_dismissed"] = pd.to_numeric(df["player_dismissed"])
df["player_dismissed"] = df.groupby("match_id")["player_dismissed"].cumsum()
```

```python
df["wickets_left"] = 10 - df["player_dismissed"]
```

```python
df
```

```python
df["crr"] = (df["current_score"]*6)/(df["balls_played"])
df = df.round({"crr" : 2})
```

```python
df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
df['runs_left'] = (df['runs_x']+1) - df['current_score']
```

```python
df
```

```python
df['rrr'] = (df['runs_left']*6)/df['balls_left']
df = df.round({"rrr" : 2})
```

```python
df
```

```python
final_df = df[['batting_team','bowling_team','runs_left','balls_left','wickets_left','runs_x','crr','rrr','result']]
```

```python
final_df
```

```python
final_df.isnull().sum()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(),['batting_team','bowling_team'])
]
,remainder='passthrough')
```

```python
X = final_df.drop("result", axis = 1)
X["rrr"].replace([np.inf, -np.inf], 0, inplace=True)
y = final_df["result"]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.15,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
```

```python
X_train
```

```python
param_grid = {
    'C': [0.01, 0.1, 1, 10],           
    'penalty': ['l1', 'l2'],           
    'solver': ['liblinear', 'lbfgs']   
}

model = LogisticRegression()


grid_search = GridSearchCV(
    model,         
    param_grid,   
    cv=5,          
    scoring='accuracy',  
    verbose=1      
)
```

```python
log_model = Pipeline(steps=[
    ('step1',trf),
    ('step2',grid_search)
])
```

```python
log_model.fit(X_train,y_train)
```

```python
print("Best Hyperparameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)
```

```python
y_pred = log_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
