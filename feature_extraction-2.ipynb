```python
import pandas as pd
import numpy as np
```

```python
import pickle
df = pickle.load(open("dataset_2.pkl" , "rb"))
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(["venue" , "city"] , axis =1 , inplace = True)
```

```python
df
```

```python
df["runs"] = pd.to_numeric(df["runs"])
df["current_score"] = df.groupby("match_id")["runs"].cumsum()
```

```python
df
```

```python
df["overs"] = df["ball"].apply(lambda x : str(x).split(".")[0])
df["ball_no"] = df["ball"].apply(lambda x : str(x).split(".")[1])
```

```python
df
```

```python
df["balls_played"] = df["overs"].astype(int) * 6 + df["ball_no"].astype(int)
```

```python
df
```

```python
df["balls_left"] = 300 - df["balls_played"]
df["balls_left"] = df["balls_left"].apply(lambda x:0 if x<0 else x)
```

```python
df
```

```python
df["player_dismissed"] = df["player_dismissed"].apply(lambda x:0 if x=="0" else 1)
df["player_dismissed"] = pd.to_numeric(df["player_dismissed"])
df["player_dismissed"] = df.groupby("match_id")["player_dismissed"].cumsum()
```

```python
df["wickets_left"] = 10 - df["player_dismissed"]
```

```python
df
```

```python
df["crr"] = (df["current_score"]*6)/(df["balls_played"])
```

```python
df
```

```python
df.info()
```

```python
df["runs"] = pd.to_numeric(df["runs"], errors="coerce")
groups = df.groupby("match_id")
match_ids = df["match_id"].unique()
last_five = []
for ids in match_ids:
    last_five.extend(groups.get_group(ids)["runs"].rolling(window=30).sum().values.tolist())
```

```python
df["last_five"] = last_five
```

```python
df
```

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
total_balls = pd.DataFrame(df.groupby("match_id").count()["ball"])
```

```python
list_drop = total_balls[total_balls["ball"] <180].index
list_drop
```

```python
final_df = final_df[~final_df['match_id'].isin(list_drop)]
final_df
```

```python
final_df = final_df[["batting_team" , "bowling_team" , "current_score" , "balls_left" , "wickets_left" , "crr" , "last_five" , "runs_x"]]
```

```python
final_df.dropna(inplace = True)
```

```python
final_df.isnull().sum()
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
from sklearn.model_selection import train_test_split
X = final_df.drop("runs_x" , axis = 1)
y = final_df["runs_x"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score , mean_absolute_error , mean_squared_error
from sklearn.model_selection import GridSearchCV
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators= 1000 ,
                          learning_rate = 0.1 ,
                          max_depth = 12))
])
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
print(np.sqrt(mean_squared_error(y_test , y_pred)))
```

```python
pickle.dump(pipe,open('pipe4.pkl','wb'))
```
